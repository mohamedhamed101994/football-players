```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
data = pd.read_csv("2022-2023 Football Player Stats.csv", encoding="latin1")
```

```python
data
```

## Issue:
#### * all columns header are concatenated on a one column header

```python
data.loc[len(data)] = data.columns  # creat a row with the same columns header name
data= data.iloc[:,0].str.split(";", expand=True)   # to split the concatenated header to many columns.
data.columns =data.iloc[2689] # to edit the data columns header again after split
data=data.drop(index=2689) # to drop the column which was created with the headers name 
```

```python
data.head()
```

```python
data.columns.rename("", inplace=True) # to edit the index col name
```

```python
df_num =data.drop(columns=["Rk","Player","Nation","Pos","Squad","Comp"]) #create a new df for numeric columns only.
df_num
```

```python
df_num = df_num.apply(pd.to_numeric) #change dtype for all numeric columns.
```

```python
df_num.dtypes
```

```python
#create the final df for the obj columns concat with the numeric columns.
df = pd.concat((data[["Rk","Player","Nation","Pos","Squad","Comp"]],df_num),axis=1)
df["Rk"]=pd.to_numeric(df["Rk"])  # change dtype for RK column to numeric.
```
